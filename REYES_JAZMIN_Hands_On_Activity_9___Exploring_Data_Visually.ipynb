**3. Discussions:**

Exploratory Data Analysis (EDA) is the process of analyzing datasets to summarize their main characteristics and gain insights before formal modeling. Visual exploration helps identify trends, relationships, and data quality issues.

Key Components:

Univariate Analysis:
Analyzing one variable at a time (e.g., histograms, bar charts).

Bivariate Analysis:
Exploring relationships between two variables (e.g., scatterplots, crosstabs).

Missing Data Analysis:
Identifying and handling missing or null values.

Time-Series Visualization:
Displaying data across time to reveal trends or seasonality.

Geospatial Visualization:
Mapping data to geographical regions to identify spatial patterns.

EDA is both an art and science combining statistical summaries with visual intuition.

Section 1: Organizing and Exploring Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample dataset
data = {
    "Product": ["Juice", "Smoothie", "Juice", "Smoothie", "Juice", "Smoothie"],
    "Region": ["North", "South", "East", "West", "North", "East"],
    "Sales": [120, 150, 90, 110, 130, 170],
    "Profit": [30, 45, 20, 25, 35, 50]
}
df = pd.DataFrame(data)

# Display the dataset
df
```

Task 1: Analyzing Variables One at a Time

```python
#Categorical Variable
# Countplot for categorical variable
sns.countplot(x="Region", data=df, palette="Set2")
plt.title("Frequency of Sales by Region")
plt.show()
```

This code creates a bar graph that shows how many sales happened in each region. It uses the Seaborn library to make the graph, with “Region” displayed on the x-axis and the number of sales on the y-axis. The palette="Set2" option is used to choose the colors of the bars, but a warning appears saying this method will not work in future versions. Finally, plt.show() is used to display the graph on the screen.

```python
#Quantitative Variable
# Histogram for quantitative variable
plt.hist(df["Sales"], bins=5, color="skyblue", edgecolor="black")
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

This markdown cell explains what the graph from the last code does. It says the bar graph shows how many sales happened in each region and also mentions the warning that the way the color palette was used will not work in future versions.

Task: <br>
Create one countplot for Product and one histogram for Profit. <br>
Compare their patterns and interpret what they mean.

```python
# Countplot for 'Product'
sns.countplot(x="Product", data=df, palette="viridis", hue="Product", legend=False)
plt.title("Frequency of Sales by Product")
plt.show()

# Histogram for 'Profit'
plt.hist(df["Profit"], bins=5, color="lightcoral", edgecolor="black")
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

**Interpretation:**

*   **Product Countplot:** The countplot shows that "Juice" and "Smoothie" have an equal frequency in the dataset, meaning there are the same number of sales records for each product.
*   **Profit Histogram:** The histogram shows the distribution of profit values. The data is somewhat spread out, with a peak in the lower profit ranges. This suggests that while sales occur for both products, the profit margins might be more varied, with a tendency towards lower profits.

**Interpretation:**

*   **Product Countplot:** Observe which product appears more frequently in the dataset. This gives an idea of the distribution of sales records between "Juice" and "Smoothie".
*   **Profit Histogram:** Examine the shape of the histogram. It shows how the profit values are distributed. Look for peaks (most frequent profit ranges) and the spread of the data.

Section 2: Relationships Between Variables

```python
# Scatterplot: Sales vs Profit
sns.scatterplot(x="Sales", y="Profit", hue="Region", data=df, palette="coolwarm", s=100)
plt.title("Relationship between Sales and Profit")
plt.show()
```

```python
# Crosstabulation example
pd.crosstab(df["Region"], df["Product"], values=df["Sales"], aggfunc="mean").fillna(0)
```

Task: <br>
Interpret whether higher sales also mean higher profit. <br>
Which region seems to perform best?

The graph shows that when sales go up, profit also goes up, meaning higher sales usually bring higher profit. Among all regions, the East region has the highest sales and profit, so it performs the best. The other regions have lower sales and lower profit, which supports the pattern that more sales lead to more profit.

Section 3: Analysis of Missing Data

```python
# Introduce missing data
df_missing = df.copy()
df_missing.loc[2, "Profit"] = np.nan
df_missing.loc[5, "Sales"] = np.nan

# Visualize missing data
sns.heatmap(df_missing.isnull(), cbar=False, cmap="Reds")
plt.title("Missing Data Visualization")
plt.show()

# Display missing summary
df_missing.isnull().sum()
```

Task: <br>
**Describe** what you observe in the missing data visualization. <br>
Which variables need attention before analysis?

The missing data visualization shows that there are empty or missing values in the Sales and Profit columns, while Product and Region have no missing data. This means that before doing any analysis, we need to focus on fixing or handling the missing values in Sales and Profit to make sure the results are accurate and complete.*

Section 4: Visualizing Time-Series Data

Task: <br>
Add a moving average line (rolling mean) to smooth fluctuations. <br>
Hint: Use ts_df["Sales"].rolling(window=3).mean().

```python
# Generate time-series example
dates = pd.date_range(start="2024-01-01", periods=10)
sales = np.random.randint(80, 200, 10)
ts_df = pd.DataFrame({"Date": dates, "Sales": sales})

# Calculate the moving average
ts_df["Sales_Rolling_Mean"] = ts_df["Sales"].rolling(window=3).mean()


# Line chart for time-series
plt.plot(ts_df["Date"], ts_df["Sales"], marker="o", color="teal", label="Sales")
plt.plot(ts_df["Date"], ts_df["Sales_Rolling_Mean"], color="red", linestyle="--", label="3-Period Rolling Mean")
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```

Section 5: Visualizing Geospatial Data

```python
import geopandas as gpd
from shapely.geometry import Point

# Sample coordinates (latitude, longitude)
coords = {
    "Branch": ["North", "South", "East", "West"],
    "Latitude": [15.5, 8.5, 13.0, 10.5],
    "Longitude": [120.9, 123.0, 125.0, 122.0],
    "Sales": [120, 150, 90, 110]
}

geo_df = pd.DataFrame(coords)
geo_df["geometry"] = [Point(xy) for xy in zip(geo_df.Longitude, geo_df.Latitude)]
gdf = gpd.GeoDataFrame(geo_df, geometry="geometry")

# Plot simple map (Philippines outline optional if available)
gdf.plot(column="Sales", cmap="Blues", legend=True, markersize=geo_df["Sales"])
plt.title("Geospatial Visualization of Sales by Branch")
plt.show()
```

This kind of map helps businesses easily see which branch is performing best and which ones are not doing as well. With this information, managers can focus on strategies like expanding successful branches or improving weaker ones to increase overall sales.
