**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()

# Customer/Teammate audience: prefer proportion/contribution (pie chart)
plt.figure(figsize=(7, 7))
plt.pie(df["Sales"], labels=df["Month"], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("viridis", len(df)))
plt.title("Customer/Teammate View: Proportion of Sales by Month", fontsize=13)
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
#Code Here
# Highlight the lowest sales month (preattentive color cue)
highlight = df["Sales"].idxmin()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Lowest Month in Sales", fontsize=13)
plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: ResolutionQ
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

The sales started steady at the beginning of the year, then experienced a drop in March because of the lower marketing spend. Thankfully, the sales recovered and showed growth after March as marketing efforts increased.


Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

**5. Supplementary Activity:**

```python
#1. Use a small dataset available online (kaggle, google dataset, etc.).
from google.colab import drive
drive.mount('/content/drive')
df  = pd.read_csv("/content/drive/MyDrive/CPE 031_REYES, JAZMIN/car_sales_data.csv")
df
```

```python
# 2. Visualize it using a bar chart.
avg_price_by_manufacturer = df.groupby('Manufacturer')['Price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(avg_price_by_manufacturer.index, avg_price_by_manufacturer.values, color='steelblue')
plt.title('Average Car Price by Manufacturer', fontsize=14)
plt.xlabel('Manufacturer', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Highlight one important data point (e.g., the highest or lowest value) using a distinct color.
plt.figure(figsize=(12, 6))
bars = plt.bar(avg_price_by_manufacturer.index, avg_price_by_manufacturer.values, color='steelblue')

highest_avg_price_manufacturer = avg_price_by_manufacturer.idxmax()
highest_avg_price_value = avg_price_by_manufacturer.max()

for bar, manufacturer in zip(bars, avg_price_by_manufacturer.index):
    if manufacturer == highest_avg_price_manufacturer:
        bar.set_color('red')
        plt.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                 f'{highest_avg_price_value:.2f}',
                 ha='center', va='bottom', color='red')

plt.title('Average Car Price by Manufacturer (Highest Average Highlighted)', fontsize=14)
plt.xlabel('Manufacturer', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Porsche is known for having the highest average car price, as shown by the highlighted bar in the chart. This sets it apart from other car brands like Ford, VW, and Toyota, whose prices are much closer to each other. The big difference in price shows that Porsche focuses on the luxury market, offering more expensive and higher-end models. This highlights how car brands use different pricing strategies based on their target customers, with Porsche positioning itself as a premium option in the market.
